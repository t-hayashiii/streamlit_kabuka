## 17.株価情報の取得 by yfinance(株価の取得)

```python
yahagi = yf.Ticker("1870.T")
yahagi
```

```python
# panadaのデータフレーム形状でデータを取得
yahagi.history(period='10d')  # 引数に取得したい日数を指定可能
# .shape属性で行列数が取得できる
```

```python
# yahagi.history(period='max')  # 最大で取得できる日数
```

```python
# 取得したい日数を変数の初期値とする
days = 10
hist = yahagi.history(period=f'{days}d')
hist
```

## 18. データ構造について

```python
hist.columns
# Dateが入っていない。なぜかコラムが2段になっている → 自動で割り振られるインデックス番号がない
# Dateがインデックスの役割をもっている
# ★複数社のデータを結合しやすい。日付をインデックスとできるから
```

```python
# hist.reset_index()  # インデックスがリセットされて、インデックスが割り振られる
```

```python
# 戸田建設の株価
hist_toda = yf.Ticker('1860.T').history(period=f'{days}d')
hist_toda.head()
```

```python
hist_toda[['Close']]
```

```python
# サンプル : データを結合
pd.concat([hist, hist_toda], axis=1).head() # データを横にくっつけることができる(日付リンク)
```

## 19. YFinanceから株価取得(データの整形)

```python
hist.head(3)
```

```python
hist.index # index列であるDateを取得
```

```python
# Date表示形式を変更
# https://www.ibm.com/docs/en/cmofm/9.0.0?topic=SSEPCD_9.0.0/com.ibm.ondemand.mp.doc/arsa0257.html
hist.index = hist.index.strftime('%d %B %Y')
```

```python
hist.head()
```

```python
hist.columns
```

```python
# 終値closeのみを使用
hist = hist[["Close"]]  
hist.columns = ['矢作']
hist.head()
```

```python
# 表に合わせて行列を入替え
hist = hist.T
hist

```

```python
# index名にNameを入れる
hist.index.name = 'Name'
hist
```
